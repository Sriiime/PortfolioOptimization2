```python
stocks = ['INND', 'LUMN', 'ASTS', 'SMMT']
```

```python
return_stocks = [0.071561, 0.032443, 0.022553, 0.016130]
```

```python
import numpy as np
```

```python
# Define the historical returns for the selected stocks
returns = np.array([return_stocks])

# Define the covariance matrix
cov_matrix = np.cov(returns.T)
```

```python
# Define the optimization function
def optimize_portfolio(weights):
    portfolio_return = np.sum(returns.mean(axis=1) * weights)
    portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return -portfolio_return / portfolio_risk
```

```python
# Define the constraints
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})  # sum of weights = 1

# Define the bounds
bounds = tuple((0, 1) for _ in range(len(stocks)))  # weights between 0 and 1
```

```python
# Initial guess for weights
init_guess = np.array([1.0/len(stocks)]*len(stocks))
```

```python
from scipy.optimize import minimize
```

```python
# Run optimization
result = minimize(optimize_portfolio, init_guess, method='SLSQP', constraints=constraints, bounds=bounds)

# Print optimized weights
print("Optimized Weights:")
print(result.x)
```

```python
#Here's the updated code:

# Calculate the optimal portfolio return
optimal_portfolio_return = np.sum(returns.mean(axis=1) * result.x)

print("Optimal Portfolio Return:", optimal_portfolio_return)
```
